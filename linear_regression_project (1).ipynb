```python
from datetime import datetime
data.Date=pd.to_datetime(data.Date)
data.info()
data['Month']=pd.DatetimeIndex(data['Date']).month
data['Year']=pd.DatetimeIndex(data['Date']).year
data['tday']=pd.DatetimeIndex(data['Date']).day
data.head()
```

```python
data_categorical=data.select_dtypes(include='object')
data_categorical
```

```python
dummies=pd.get_dummies(data_categorical['Seasons'])
```

```python
dummies
```

```python
dummies_2=pd.get_dummies(data_categorical['Functioning Day'])
dummies=pd.concat([dummies,dummies_2],axis='columns')
dummies
```

```python
dummies_3=pd.get_dummies(data_categorical['Holiday'])
dummies=pd.concat([dummies,dummies_3],axis='columns')
dummies
```

```python
data=pd.concat([data,dummies],axis='columns')
data.head()
```

```python
data.drop({'Date','Seasons','Holiday','Functioning Day'},inplace=True,axis='columns')
data.head()
```

```python
#Spliting dataframe into train and test 
from sklearn.model_selection import train_test_split
train,test=train_test_split(data,test_size=0.25)


```

```python
train
```

```python
trainx=train.drop('Rented Bike Count',axis='columns')
trainy=train['Rented Bike Count']
testx=test.drop('Rented Bike Count',axis='columns')
testy=test['Rented Bike Count']
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(trainx,trainy)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
df_pred=lr.predict(testx)
```

```python
print("The model's metrics values are: ")
print("Accuracy: ", lr.score(testx,testy))
print("Mean Absolute Error: ",mean_absolute_error(df_pred , testy))
print("Mean Squared Error: " ,mean_squared_error(df_pred,testy))
print("R-squared score: ", r2_score(testy,df_pred))
```
